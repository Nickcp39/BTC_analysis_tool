```python
# Cell 1: 基本库 & 路径
from pathlib import Path
import pandas as pd

# 把这两行改成你的实际文件名/路径（建议把 .ipynb 放到和数据同一文件夹，路径就只写文件名）
f_2021 = Path("btc_2021.xlsx")
f_2025 = Path("btc_2025.xlsx")   # 例如 8 月份的

assert f_2021.exists() and f_2025.exists(), "请检查两个 Excel 路径是否正确"
```

```python
# ==== 配置：改这里的文件名即可 ====
from pathlib import Path
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

F_21 = Path("btc_2021.xlsx")  # 2019/2020/2021 混在一起也没问题
F_25 = Path("btc_2025.xlsx")  # 2024/2025 混在一起也没问题

# 分析窗口（锚点前后各取多少天）
PRE_DAYS, POST_DAYS = 120, 120

# 锚点设定：2025 固定用 8/15（你的“针尖”假设）
ANCHOR_2025 = pd.Timestamp("2025-08-15")
END_2025    = pd.Timestamp("2025-08-22")   # 你现在的数据大概到这一天

# ==== 工具函数 ====
def read_two_col_excel(fp: Path) -> pd.DataFrame:
    assert fp.exists(), f"未找到文件：{fp}"
    df_raw = pd.read_excel(fp, engine="openpyxl", header=None)
    # 如果是“横着两行”的格式（第1行时间，第2行价格），自动转置
    if df_raw.shape[0] <= 3 and df_raw.shape[1] > df_raw.shape[0]:
        df_raw = df_raw.T

    df = df_raw.iloc[:, :2].copy()
    df.columns = ["ts", "price"]
    df["ts"] = pd.to_datetime(df["ts"], errors="coerce")
    df["price"] = pd.to_numeric(df["price"], errors="coerce")
    df = df.dropna(subset=["ts", "price"]).sort_values("ts")
    df = df.drop_duplicates(subset=["ts"], keep="last")
    return df

def to_daily_series(df: pd.DataFrame, start=None, end=None) -> pd.Series:
    """转为日频收盘（last），并向前填充以获得连续日期索引。"""
    if start is None: start = df["ts"].min().normalize()
    if end   is None: end   = df["ts"].max().normalize()
    daily = df.set_index("ts")["price"].resample("D").last()
    daily = daily.reindex(pd.date_range(start, end, freq="D")).ffill()
    daily.index.name = "ts"
    return daily

def make_rel_curve(daily: pd.Series, anchor_date: pd.Timestamp, pre_days: int, post_days: int):
    """围绕锚点截取窗口，并做归一化（锚点价=1.0），同时给出锚点价。"""
    anchor_date = pd.Timestamp(anchor_date).normalize()
    lo = max(anchor_date - pd.Timedelta(days=pre_days), daily.index.min())
    hi = min(anchor_date + pd.Timedelta(days=post_days), daily.index.max())
    win = daily.loc[lo:hi].copy()
    # 锚点价：若无当日，就取锚点日前最近可得值
    anchor_price = float((win.loc[anchor_date] if anchor_date in win.index else win.loc[:anchor_date].iloc[-1]))
    rel_day = (win.index - anchor_date).days
    norm = win / anchor_price
    dd_pct = (norm - 1.0) * 100.0
    return (pd.DataFrame({"rel_day": rel_day, "norm": norm.values, "dd_pct": dd_pct.values}, index=win.index),
            anchor_price)

# ==== 读取与预处理 ====
df21 = read_two_col_excel(F_21)
df25 = read_two_col_excel(F_25)

daily21_full = to_daily_series(df21)
daily25_full = to_daily_series(df25)

# 只取 2021 年整年 & 2025 年到 8/22
daily21 = daily21_full.loc["2021-01-01":"2021-12-31"]
daily25 = daily25_full.loc["2025-01-01":END_2025]

assert not daily21.empty, "2021 年数据为空，请检查 btc_2021.xlsx"
assert not daily25.empty, "2025 年数据为空，请检查 btc_2025.xlsx"

# 2021 锚点：当年最高价日期（若想手动指定，可改成 pd.Timestamp('2021-11-10')）
peak_2021_date = daily21.idxmax()

# 2025 锚点：固定 8/15
rel21, anchor21 = make_rel_curve(daily21, peak_2021_date, PRE_DAYS, POST_DAYS)
rel25, anchor25 = make_rel_curve(daily25, ANCHOR_2025, PRE_DAYS, POST_DAYS)

# ==== 拟合评估 ====
# 锚点前（-N..0）比较：看爬坡相似度（归一化）
pre_21 = rel21.loc[rel21["rel_day"] <= 0].set_index("rel_day")["norm"]
pre_25 = rel25.loc[rel25["rel_day"] <= 0].set_index("rel_day")["norm"]
pre_idx = pre_21.index.intersection(pre_25.index)
pre_corr = np.corrcoef(pre_21.loc[pre_idx], pre_25.loc[pre_idx])[0,1] if len(pre_idx) >= 3 else np.nan
pre_rmse = float(np.sqrt(np.mean((pre_21.loc[pre_idx].values - pre_25.loc[pre_idx].values)**2))) if len(pre_idx) >= 3 else np.nan

# 锚点后（0..+N）比较：用回撤轨迹（%）避免绝对价差影响
post_21 = rel21.loc[rel21["rel_day"] >= 0].set_index("rel_day")["dd_pct"]
post_25 = rel25.loc[rel25["rel_day"] >= 0].set_index("rel_day")["dd_pct"]
post_idx = post_21.index.intersection(post_25.index)
post_corr = np.corrcoef(post_21.loc[post_idx], post_25.loc[post_idx])[0,1] if len(post_idx) >= 3 else np.nan
post_rmse = float(np.sqrt(np.mean((post_21.loc[post_idx].values - post_25.loc[post_idx].values)**2))) if len(post_idx) >= 3 else np.nan

# ==== 画图（各一张）====
plt.figure(figsize=(10,4))
plt.plot(rel21["rel_day"], rel21["norm"], label=f"2021（锚点 {peak_2021_date.date()}）")
plt.plot(rel25["rel_day"], rel25["norm"], label=f"2025（锚点 {ANCHOR_2025.date()}）")
plt.axvline(0, linestyle="--"); plt.xlabel("相对天数（锚点=0）"); plt.ylabel("归一化价格")
plt.title("锚点归一化价格（锚点=1.0）"); plt.legend(); plt.tight_layout(); plt.show()

plt.figure(figsize=(10,4))
plt.plot(post_21.index, post_21.values, label="2021：锚点后回撤轨迹（%）")
plt.plot(post_25.index, post_25.values, label="2025：锚点后回撤轨迹（%）")
plt.axvline(0, linestyle="--"); plt.xlabel("相对天数（锚点=0）"); plt.ylabel("相对锚点涨跌（%）")
plt.title("锚点后回撤轨迹对比"); plt.legend(); plt.tight_layout(); plt.show()

print("—— 锚点信息 ——")
print(f"2021 锚点日期：{peak_2021_date.date()}，锚点价：{anchor21:,.2f}")
print(f"2025 锚点日期：{ANCHOR_2025.date()}（假设） ，锚点价：{anchor25:,.2f}")
print("\n—— 拟合评估（重叠样本） ——")
print(f"锚点前（-N..0）相关系数 r = {pre_corr:.4f}，RMSE = {pre_rmse:.4f}")
print(f"锚点后（0..+N）回撤轨迹相关 r = {post_corr:.4f}，RMSE(百分比点) = {post_rmse:.4f}")

# 可选：保存图片
out1 = Path("normalized_anchor_curve.png")
out2 = Path("post_peak_drawdown.png")
plt.figure(figsize=(10,4))
plt.plot(rel21["rel_day"], rel21["norm"], label=f"2021（锚点 {peak_2021_date.date()}）")
plt.plot(rel25["rel_day"], rel25["norm"], label=f"2025（锚点 {ANCHOR_2025.date()}）")
plt.axvline(0, linestyle="--"); plt.xlabel("相对天数（锚点=0）"); plt.ylabel("归一化价格")
plt.title("锚点归一化价格（锚点=1.0）"); plt.legend(); plt.tight_layout(); plt.savefig(out1, dpi=160); plt.close()

plt.figure(figsize=(10,4))
plt.plot(post_21.index, post_21.values, label="2021：锚点后回撤轨迹（%）")
plt.plot(post_25.index, post_25.values, label="2025：锚点后回撤轨迹（%）")
plt.axvline(0, linestyle="--"); plt.xlabel("相对天数（锚点=0）"); plt.ylabel("相对锚点涨跌（%）")
plt.title("锚点后回撤轨迹对比"); plt.legend(); plt.tight_layout(); plt.savefig(out2, dpi=160); plt.close()

print("已保存图片：", out1.resolve())
print("已保存图片：", out2.resolve())
```
